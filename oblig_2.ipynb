Rotete notater der jeg prøvde å faktorisere den deriverte for å finne mulig nullpunkt

Ettersom at det ikke virket mulig så deriverte jeg den deriverte, for å finne nullpunkt via newtons metode

Inkluderte ikke notater der jeg regnet ut mulig nullpunkt med newtons metode på grunn av at jeg deriverte feil første gangen.
Etter feilen var rettet estimerte jeg at x1 = 1.67 når x0 var 1.5 med vanlig kalkulator, som stemte med visualisering av grafen via geogebra.
Jeg implementerte da funksjonene og regnet ut svaret numerisk.

Utregning av første og andre deriverte av f

```python
import numpy as np
import matplotlib.pyplot as plt

#n antall iterasjoner for å finne x
iterations = 1000

#gjettet startverdi
x = 1

#funksjon F
def orig_F(x):
    return np.exp(-x/4.0) * np.arctan(x) 

#den første deriverte funksjonen av F
def fn(x):
    return np.arctan(x) - 4/(x**2 +1) 

#den andre deriverte funksjonen av F, f'
def der_fn(x):
    return ((8 * x) / (x**2 + 1)**2) + (1 / (x**2 + 1))

#newtons metode
for i in range(iterations):
    #xn+1 = x - f/f'
    x = x - fn(x)/der_fn(x)

#x etter newtons metode
svaret = x
print("x = " + str(svaret))
```

```python
print("toppunkt F(" + str(svaret) + ") = " + str(orig_F(svaret)))
print("når f(" + str(svaret) + ") = " + str(fn(svaret)))
```

```python

x = np.linspace(-10,10,5000)

#funksjon F
f = orig_F(x)

#den deriverte funksjonen av f
d = fn(x)

#for å vise krysningen av x aksen til den deriverte
x_axis = [-10,10]
y_axis = [0,0]

#rød linje er F
plt.plot(x,f,'r-',linewidth = 1)

#blå linje er den deriverte, f 
plt.plot(x,d,'b-',linewidth = 1)

#verdien til xn etter newtons metode der den deriverte er = 0
plt.plot(svaret,0, marker = 'o')

#toppunktet til funksjon F
plt.plot(svaret,orig_F(svaret), marker = 'o')

#aksene
plt.plot(x_axis,y_axis,'k',linewidth = 1)
plt.plot(y_axis,x_axis,'k',linewidth = 1)

plt.axis([-5.0,10.0,-5.0,5.0])
plt.show
```
