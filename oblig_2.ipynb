```python
import numpy as np
import matplotlib.pyplot as plt

iterations = 1000

x = 1

def orig_f(x):
    return np.exp(-x/4.0) * np.arctan(x) 

def fn(x):
    return np.arctan(x) - 4/(x**2 +1) 

def der_fn(x):
    return ((8 * x) / (x**2 + 1)**2) + (1 / (x**2 + 1))

for i in range(iterations):
    x = x - fn(x)/der_fn(x)

print(x)
```

```python

x = np.linspace(-10,10,5000)
f = orig_f(x)
d = fn(x)

x_axis = 0
y_axis = 1

plt.plot(x,f)
plt.plot(x,d)
plt.plot(x_axis,y_axis)
plt.axis([-10,10,-10,10])
plt.show
```
