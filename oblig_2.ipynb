```python
import numpy as np
import matplotlib.pyplot as plt

#xn antall iterasjoner for å finne xn
iterations = 1000

#gjettet startverdi
x = 1

#funksjon F
def orig_f(x):
    return np.exp(-x/4.0) * np.arctan(x) 

#den første deriverte funksjonen av F
def fn(x):
    return np.arctan(x) - 4/(x**2 +1) 

#den andre deriverte funksjonen av F
def der_fn(x):
    return ((8 * x) / (x**2 + 1)**2) + (1 / (x**2 + 1))

#newtons metode
for i in range(iterations):
    x = x - fn(x)/der_fn(x)

#xn etter newtons metode
svaret = x
print("Xn = " + str(x))
```

```python
print("toppunkt til F = " + str(orig_f(svaret)))
```

```python

x = np.linspace(-10,10,5000)

#funksjon F
f = orig_f(x)

#den deriverte funksjonen av F
d = fn(x)

#for å vise krysningen av x aksen til den deriverte
x_axis = [-10,10]
y_axis = [0,0]

#rød linje er F
plt.plot(x,f,'r-',linewidth = 1)

#blå linje er den deriverte
plt.plot(x,d,'b-',linewidth = 1)

#verdien til xn etter newtons metode der den deriverte er = 0
plt.plot(svaret,0, marker = 'o')

#toppunktet til funksjon F
plt.plot(svaret,orig_f(svaret), marker = 'o')

plt.plot(x_axis,y_axis,'k',linewidth = 1)
plt.plot(y_axis,x_axis,'k',linewidth = 1)

plt.axis([-5,10,-5,5])
plt.show
```
